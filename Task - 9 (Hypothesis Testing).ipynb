# Hypothesis Testing

### What is Hypothesis testing?

Validating our hypothesis or claims with help of data is called as hypothesis testing. Previoulsy, we have inference where we infer about some kind of an observation. But, in hypothesis testing we test that with some certain level of confidence interval and cateogorize it into null or alternate hypothesis

Hypothesis Testing starts with the formulation of these two hypotheses: <br>
>• __Null hypothesis (H₀)__: The status quo<br>
• __Alternate hypothesis (H₁)__: The challenge to the status quo (__The BOLD claim__)<br>


__Example:__ _Suppose, there are two people and they are discussing about the price of an ice cream. The first one days that its 90 rs. While the second one did not agree to that.  <br>


>• __The null hypothesis__ is the belief that an ice cream costs 90 rs. It states that there is no change or no
difference in the situation. So,
the null hypothesis claims that it is equal to 90, it is denoted by __H0__ <br>

>• __The alternate hypothesis__ or research hypothesis as it is also called, is the Bold claim that opposes the
null hypothesis. In our example the second friend did not agreed that the ince cream price is 90, and you would try to prove this. So, the alternate hypothesis is an assumption that competes
with the null hypothesis. Alternate Hypothesis is denoted by __H1__.

### Outcome of the hypothesis testing:


> In our example, we can see that friend 1 collected a sample of 2 ice creams and calculated their mean price which found to be 120!!.<br>
__In this case, What would be the outcome in terms of hypothesis testing?__ <br>

> * If the calculated sample mean is not equal to the assumption made, it means that this rejects the null hypothesis. So, we can decide that there is enough evidence to support the alternate hypothesis, and to
conclude that the population price is not 90.<br>

>* On the other hand, if suppose the mean of the sample turns out to be 90, it means that there is not enough evidence to support
the alternate hypothesis. Note that this does not mean that the price is 90, it just means
that there is not enough evidence to conclude that it is not 90. In other words, we cannot accept the null
hypothesis; we can only fail to reject it.<br>

>Therefore, in hypothesis testing, if there is sufficient evidence to support the alternate hypothesis, you __reject
the null hypothesis__; <br> if there is not sufficient evidence to support the alternate hypothesis, you __fail to
reject the null hypothesis__. <br>

>So, you should never say that you “accept” the null hypothesis

### Rules
• The null hypothesis __H0__ always has the following signs: __= OR ≤ OR ≥__ <br>
• The alternate hypothesis H1 always has the following signs: ≠ OR > OR < <br>

The understanding of the hypothesis testing is made with many different techniques.. <br>
Some of the techniques are:- <br>
* Z test
* p value test
* t test
* chi square testing

__T-distribution (or Student T distribution)__: It is similar to the normal distribution in many cases. For example,
it is symmetrical about its central tendency. However, it is usually shorter than the normal distribution and has a
flatter tail, which would eventually mean that it has a larger standard deviation.<br>

The following graph shows t-distribution vs normal distribution<br>


__T-Distribution characteristics:__
>* Better to use T - Distribution when you have a sample size less than 30
>* Beyond smaple size of 30, the t-distribution becomes normal distribution.<br>
>* T - Distribution is distinguished by its degrees of freedom, where the degrees of freedom for the corresponding T - Distribution is n -1. (For example, if your sample size is 10, you use a t-distribution with 10 -1 or 9 degrees of freedom, denoted t9.) <br>
>* The larger the sample size becomes, the larger will be the degree of freedom. So, our distribution looks like normla distribution.
>* Used when we dont have standard distribution

# Steps in Hypothesis testing

 __Step 1:__ Understanding the scenario and plot the Null hypothesis and Alternate hypothesis

__Step 2:__ Collecting a sample of size N and computing its mean.

__Step 3:__ Calculate the test statistic value t.
> The formula for calculating t is given as, <br>
$$t = \frac{\bar{X}-\mu}{\frac{s}{\sqrt[2]{N}}}$$
>Where, <br>
* t = Test value 
* $\bar{X}$ = Mean of the sample
* $\mu$ = Mean of the population
* s = standard deviation of the sample
* N = Sample size

__Step 4:__ We should chooce the significance level of our hypothesis which is denoted by $\alpha$. The lower the significance level, the more confident you can be in replicating your results. </p>

> Lwt us assume the confidence level is 95% i.e. 0.95 <br>
$\alpha$ = 1 - confidence level <br>
Therefore $\alpha$ = 0.05

__Step 5:__ Deciding the type of test.<br>
We have to decide the type of test depending on the context of hypothesis. The tests are,<br>
* Two-tailed test
* One-tailed test <br>
_You can tell the type of the test and the position of the critical region on the basis of the __‘sign’__ in the
alternate hypothesis._
* $\neq$ in H₁ $\rightarrow$ Two-tailed test $\rightarrow$ Rejection region on both sides of distribution
* < in H₁ $\rightarrow$ Lower-tailed test $\rightarrow$ Rejection region on left side of distribution
* \> in H₁ $\rightarrow$ Upper-tailed test $\rightarrow$ Rejection region on right side of distribution<br>

# Case 1

_Pista House selling Hyderabadi Chicken Dum biryani claims that each parcel packet has 500 grams of biryani (also mentioned on the label of packet). You are sceptic of their claims and believe that on average each packet does not contain 500 grams of biryani. How do you prove your claim?_  

<font color = blue>__Solution__:</font>
Let us follow a step by step approach to solve these type of problems. <br>

 __Step 1:__ Understanding the scenario and plot the Null hypothesis and Alternate hypothesis. For our case 1, <br>
> Null Hypothesis __$H_0$__ (Status Quo) is that the weight of the biriyani packet is 500 gms.<br>
                        $$H_0 : \bar{X} = 500$$
> Alternate Hypothesis __$H_1$__ (Bold Claim) is that the average weight of biriyani packet is not 500 gms. <br>
$$H_1 : \bar{X} \neq 500$$

__Step 2:__ Collecting a sample of size N and computing its mean.<br>
> In our example, let us collect 10 biriyani packets to form a sample.<br>
Sample size N = 10 <br>
Let the sample be, $$ [490, 420, 470, 500, 495, 496, 496, 498, 508, 480] $$

```python
# Let us import some useful modules into the environment.
import matplotlib.pyplot as plt
import numpy as np
import math
from scipy.stats import t
```

```python
pop_mean = 500 #Given
sample = [490,420,470,500,495,496,496,498,508,480]
N = len(sample)
sample_mean = sum(sample)/N
print("The mean of the sample is {}".format(sample_mean))
```

calculate t value

```python
# let us define a function for calculating t value
def t_value(sample_size, sample_mean, pop_mean, sample_std):
    numerator = sample_mean - pop_mean
    denomenator = sample_std / sample_size**0.5
    return numerator / denomenator
```

```python
# Let us find the standard deviation of the sample s.
sample_std = np.std(sample)
print("The standard deviation s of the sample of size N is {}". format(sample_std))
```

```python
t_val = t_value(N,sample_mean,pop_mean,sample_std)
print("The t static value for the given case is {}".format(t_val))
```

significance level

```python
confidence_interval = 0.95
alpha = 1 - confidence_interval
```

From step 1, the alternate hypothesis $H_1$ has a $\neq$ and we should perform a two tailed test.

```python
# To find the t value of a two tailed test,
t_critical = t.ppf(1-alpha/2,df = N-1)
print("The t critical value for the given data is {}".format(t_critical))
```

```python
# interpret via critical value
if abs(t_val) >= t_critical:
    print("Reject the null hypothesis.")
else:
    print('Fail to Reject the null hypothesis.')
```

Let us plot the distribution with the t critical value and the obtained t static value to visualise the hypothesis output.

```python
# Ploting the sampling distribution with rejection regions

# Defining the x minimum and x maximum
x_min = 400
x_max = 600
df = N-1
# Defining the sampling distribution mean and sampling distribution std
mean = pop_mean
std = sample_std

# Ploting the graph and setting the x limits
x = np.linspace(x_min, x_max, 100)
y = t.pdf(x, df,mean, std)
plt.figure(figsize = (15,10))

# Computing the left and right critical values (Two tailed Test)
t_critical_left = pop_mean - (t_critical * std)
t_critical_right = pop_mean + (t_critical * std)

# Shading the left rejection region
x1 = np.linspace(x_min, t_critical_left, 100)
y1 = t.pdf(x1,df, mean, std)
plt.fill_between(x1, y1, color='red')
plt.scatter(x1,y1)

# Shading the right rejection region
x2 = np.linspace(t_critical_right, x_max, 100)
y2 = t.pdf(x2,df, mean, std)
plt.fill_between(x2, y2, color='red')
plt.scatter(x2,y2)
plt.scatter(sample_mean, 0)
plt.annotate("Sample_mean", (sample_mean-20, 0.0007))
plt.axvline(mean, color='green')
plt.annotate("pop_mean", (mean+1, 0.01))


plt.plot(x, y)
plt.show()
# Ploting the sample mean and concluding the results 
```

So, to conclude in our case 1, we failed to gather enough evidence to reject the null hypothesis.
